## Desafios

### Adicionar a coluna de região

```python
base_atendimento_formatada['Regiao'] = base_atendimento_formatada.index
```

```python
base_atendimento_formatada.head(5);
```

```python
def verifica_regiao(linha): 
  if linha[0] == '1':
    return "Norte"
  if linha[0] == '2':
    return "Nordeste"
  if linha[0] == '5':
    return "Centro-Oeste"
  if linha[0] == '3':
    return "Sudeste"
  if linha[0] == '4':
    return "Sul"
```

```python
base_atendimento_formatada.apply(lambda linha: verifica_regiao(linha['Regiao']),axis=1)
```

```python
base_atendimento_formatada['Regiao'] = base_atendimento_formatada.apply(lambda linha: verifica_regiao(linha['Regiao']),axis=1)
base_atendimento_formatada
```
