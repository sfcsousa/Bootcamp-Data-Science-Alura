# Aula 02 - Primeiras Visualizações de dados

```python
base_atendimento = pd.read_csv("/content/valor_total_ano_mes_atendimento_UF.csv", encoding='ISO-8859-1',engine='python', skiprows=3, skipfooter=12, sep=';', thousands='.',decimal=',')
pd.options.display.float_format = "{:.2f}".format
base_atendimento
```

```python
base_atendimento[['2010/Ago','2010/Set']].tail()
```

```python
base_atendimento.describe()
```

```python
base_atendimento.mean()
```

```python
colunas_usaveis = base_atendimento.mean().index.tolist()
colunas_usaveis
```

```python
colunas_usaveis.insert(0,'Unidade da Federação')
```

```python

base_atendimento_formatada = base_atendimento[colunas_usaveis]
base_atendimento_formatada
```

```python
base_atendimento_formatada = base_atendimento_formatada.set_index("Unidade da Federação")
base_atendimento_formatada.tail()
```

```python
base_atendimento_formatada['2008/Ago']
```

```python
base_atendimento_formatada.loc['12 Acre']
```

```python
base_atendimento_formatada.T.tail()
```

```python
base_atendimento_formatada.T.plot(figsize=(12,6))
```

```python
base_atendimento_formatada = base_atendimento_formatada.drop("Total",axis=1) ## eixo da linha
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
ax = base_atendimento_formatada.T.plot(figsize=(12,6))
handles, labels = ax.get_legend_handles_labels()
plt.legend(handles,labels,bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
ax.set_title("Evolução de gastos por atendimento entre agosto de 2007 e dezembro de 2020")
plt.show()
```

## Desafios 

### Reposicionar a legenda do grafico gerado de forma mais adequada

```python
ax = base_atendimento_formatada.T.plot(figsize=(12,6))
handles, labels = ax.get_legend_handles_labels()
plt.legend(handles,labels,bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
ax.set_title("Evolução de gastos com atendimentos entre agosto de 2007 e dezembro de 2020")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.show()
```

### Plotar o grafico selecionando apenas 5 estados. Hint, os estados são as colunas quando transposto

```python
ax = base_atendimento_formatada.T.plot(figsize=(13,8))
handles, labels = ax.get_legend_handles_labels()
plt.legend(handles,labels,bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
ax.set_title("Evolução de gastos com atendimentos entre agosto de 2007 e dezembro de 2020")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.show()
```

```python
estados = ["12 Acre","15 Pará","42 Santa Catarina","33 Rio de Janeiro","35 São Paulo"]
estados
```

```python
ax = base_atendimento_formatada.T.plot(figsize=(12,6),y=estados)
handles, labels = ax.get_legend_handles_labels()
plt.legend(handles,labels,bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
ax.set_title("Evolução de gastos com atendimentos entre agosto de 2007 e dezembro de 2020")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.show()
```
