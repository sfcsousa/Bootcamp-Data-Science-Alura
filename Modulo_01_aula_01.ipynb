```python
ax = dados.plot(x="Unidade da Federação", y="2008/Ago",kind="bar", figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor por unidade da federação")
plt.show()
```

## Desafios

### Escolher um titulo adequado ao gráfico

```python
ax = dados.plot(x="Unidade da Federação", y="2008/Ago",kind="bar", figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valores repassados aos estados referentes a internações realizadas em Agosto de 2008")
plt.show()
```

### Analise gráfica para um mês mais recente

```python
ax = dados.plot(x="Unidade da Federação", y=["2020/Jun","2020/Jul"],kind="bar", figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valores repassados aos estados referentes a internações realizadas em Julho de 2020")
plt.show()
```

```python
ax = dados.plot(x="Unidade da Federação", y=["2020/Jan","2020/Fev","2020/Mar"],kind="bar", figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valores repassados aos estados referentes a internações realizadas no 1o trimestre de 2020")
plt.show()
```
